# Loan Prediction using Decision Tree


```python
# Import required libraries
import pandas as pd
import numpy as np
```

```python
# Load the dataset
data = pd.read_csv('data.csv')
data.head()
```

```python
# Drop Loan_ID column (identifier, not useful for prediction)
data = data.drop('Loan_ID', axis=1)
```

```python
# Identify categorical and numerical columns
cat_cols = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']
num_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']
```

```python
# Handle missing values in categorical columns using mode
for col in cat_cols:
    data[col] = data[col].fillna(data[col].mode()[0])
```

```python
# Handle missing values in numerical columns using median
for col in num_cols:
    data[col] = data[col].fillna(data[col].median())
```

```python
# Encode ordinal categorical column (Dependents)
from sklearn.preprocessing import LabelEncoder
le_dep = LabelEncoder()
data['Dependents'] = le_dep.fit_transform(data['Dependents'])
```

```python
# One-Hot Encode nominal categorical columns
data = pd.get_dummies(
    data,
    columns=['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area'],
    drop_first=True
)
```

```python
# Separate features and target variable
X = data.drop('Loan_Status', axis=1)
Y = data['Loan_Status']
```

```python
# Encode target variable (Loan_Status)
le_target = LabelEncoder()
Y = le_target.fit_transform(Y)
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
# Train Decision Tree model
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini', random_state=42)
dt.fit(X_train, Y_train)
```

```python
# Make predictions
Y_pred = dt.predict(X_test)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score, classification_report
print('Accuracy Score:', accuracy_score(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```
