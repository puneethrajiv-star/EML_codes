#Implement linear regression algorithm on a given dataset

```python
# ==========================================
# Linear Regression Implementation
# ==========================================

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder

# ------------------------------------------
# Step 1: Load the dataset
# ------------------------------------------
data = pd.read_csv('data.csv')

# ------------------------------------------
# Step 2: Preprocessing
# ------------------------------------------
# Drop identifier column
data = data.drop('Loan_ID', axis=1)

# Categorical and numerical columns
cat_cols = ['Gender', 'Married', 'Dependents',
            'Education', 'Self_Employed', 'Property_Area']

num_cols = ['ApplicantIncome', 'CoapplicantIncome',
            'LoanAmount', 'Loan_Amount_Term', 'Credit_History']

# Handle missing values
for col in cat_cols:
    data[col] = data[col].fillna(data[col].mode()[0])

for col in num_cols:
    data[col] = data[col].fillna(data[col].median())

# Encode ordinal column
le = LabelEncoder()
data['Dependents'] = le.fit_transform(data['Dependents'])

# One-hot encode nominal categorical columns
data = pd.get_dummies(
    data,
    columns=['Gender', 'Married', 'Education',
             'Self_Employed', 'Property_Area'],
    drop_first=True
)

# ------------------------------------------
# Step 3: Separate input and output
# ------------------------------------------
X = data.drop('Loan_Status', axis=1)
Y = data['Loan_Status']

# Encode target variable
le_target = LabelEncoder()
Y = le_target.fit_transform(Y)

# ------------------------------------------
# Step 4: Train-test split
# ------------------------------------------
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

# ------------------------------------------
# Step 5: Train Linear Regression model
# ------------------------------------------
model = LinearRegression()
model.fit(X_train, Y_train)

# ------------------------------------------
# Step 6: Predict on test data
# ------------------------------------------
Y_pred = model.predict(X_test)

# ------------------------------------------
# Step 7: Model Evaluation
# ------------------------------------------
print("Mean Squared Error:", mean_squared_error(Y_test, Y_pred))
print("R2 Score:", r2_score(Y_test, Y_pred))
```
