```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans,AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
data=pd.read_csv('iris.csv')
```

```python
print(data.columns)
```

```python
x=data.drop('species',axis=1)
print(x.columns)
```

```python
print(x.head())
```

```python
scaler=StandardScaler()
x_scale=scaler.fit_transform(x)
#print(x_scale)
```

#apply Kmeans

```python
kmeans=KMeans(n_clusters=3)
kmeans_labels=kmeans.fit_predict(x)
```

#Apply Agglomerative Clustering

```python
agg=AgglomerativeClustering(n_clusters=3)
agg_labels=agg.fit_predict(x_scale)
```

#evaluate the performance

```python
kmeans_score=silhouette_score(x_scale,kmeans_labels)
agg_score=silhouette_score(x_scale,agg_labels)
print(kmeans_score,agg_score)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))


plt.subplot(1,2,1)
plt.scatter(x_scale[:,0],x_scale[:,1],c=kmeans_labels,label='kmeans')

plt.subplot(1,2,2)
plt.scatter(x_scale[:,0],x_scale[:,1],c=agg_labels,label='agg')
plt.legend()
plt.show()
```
