```python
# Declaring variables of different data types
x = 10              # Integer value
pi = 3.14           # Floating point value
name = "AI"         # String value
is_active = True    # Boolean value
```

```python
# Creating basic Python data structures
numbers = [1, 2, 3]                     # List
point = (10, 20)                        # Tuple
student = {"name": "Ravi", "age": 20}  # Dictionary
```

```python
# Conditional statement to check pass/fail
marks = 75
if marks >= 50:
    print("Pass")      # Executed if condition is True
else:
    print("Fail")      # Executed if condition is False
```

```python
# For loop to print numbers from 0 to 4
for i in range(5):
    print(i)
```

```python
# Defining a user-defined function
def add(a, b):
    return a + b      # Returns the sum of a and b
```

```python
# List comprehension to generate squares of numbers
squares = [x**2 for x in range(5)]
squares
```

```python
# Lambda function for addition (anonymous function)
add = lambda a, b: a + b
add(3, 4)             # Function call
```

```python
# Using zip() and enumerate() functions
names = ["A", "B"]
marks = [80, 90]

# zip() pairs elements from two lists
for n, m in zip(names, marks):
    print(n, m)

# enumerate() gives index and value
for idx, val in enumerate(names):
    print(idx, val)
```

```python
# Importing NumPy library
import numpy as np
```

```python
# Creating a NumPy array
arr = np.array([1, 2, 3, 4])
arr
```

```python
# Checking shape and number of dimensions of array
arr.shape     # Returns (rows, columns)
arr.ndim      # Returns number of dimensions
```

```python
# Performing array operations
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

a + b         # Element-wise addition
a.mean()      # Mean of elements
a.sum()       # Sum of elements
```

```python
# Demonstrating axis concept in NumPy
A = np.array([[1, 2],
              [3, 4]])

A.sum(axis=0)   # Column-wise sum
A.sum(axis=1)   # Row-wise sum
```

```python
# Matrix multiplication using @ operator
B = np.array([[5, 6],
              [7, 8]])

A @ B
```

```python
# Generating random numbers
np.random.rand(5)                 # Random floats between 0 and 1
np.random.randint(1, 10, size=5)  # Random integers
```

```python
# Demonstrating broadcasting
arr = np.array([1, 2, 3])
arr + 10      # Adds 10 to each element
```

```python
# Importing Pandas library
import pandas as pd
```

```python
# Creating a Pandas Series
s = pd.Series([10, 20, 30])
s
```

```python
# Creating a Pandas DataFrame
df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Age": [20, 21, 22],
    "Marks": [85, 90, 88]
})
df
```

```python
# Exploring the DataFrame
df.head()      # Displays first few rows
df.info()      # Displays DataFrame structure
df.describe()  # Statistical summary
```

```python
# Column selection
df["Marks"]                # Single column
df[["Age", "Marks"]]       # Multiple columns
```

```python
# Using loc for label-based selection and update
df.loc[0]                               # Single row
df.loc[0:2, ["Name", "Marks"]]         # Rows and columns
df.loc[df["Marks"] > 85, ["Name", "Marks"]]
df.loc[df["Name"] == "A", "Marks"] = 95
df
```

```python
# Using iloc for index-based selection
df.iloc[0]
df.iloc[0:2]
df.iloc[0:2, 1:3]
df.iloc[[0, 2], [0, 2]]
```

```python
# Sorting and unique operations
df.sort_values(by="Marks", ascending=False)
df["Age"].unique()
df["Age"].value_counts()
```

```python
# Handling missing values
df.isnull()     # Check for missing values
df.dropna()     # Drop missing values
df.fillna(0)    # Fill missing values
```

```python
# List comprehension to generate multiples of 5
five = [x * 5 for x in range(1, 11)]
print(five)
```
